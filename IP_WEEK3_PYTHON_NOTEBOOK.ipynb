            DATA CLEANING

```python
#we combine the telcom dataset since they contain  similar column names
df_combined = pd.concat([df2,df3,df4], axis=0)
df_combined
```

```python
#trying to get an insight of the combined dataset,
df_combined.info()
```

```python
#From our insight of the combined tables we notice that the Date_Time datatype has been put as object, 
# to avoid problems during calculation we should chage it to datetime format
#we then check to see if it has been updated
df_combined['Date_Time'] = pd.to_datetime(df_combined['Date_Time'])
df_combined.info()
```

```python
# basing on our target questions we check the combined tables to observe if there is a column we can do without
df_combined
```

```python
#By having a brief insight on the combined table we notice that the countryA an CountryB column
# have similar content so we can do without it
df_combined.drop(['Country_A', 'Country_B'], axis=1)
```

```python
#From our data understanding stage above we notice that we have duplicates in our tables 
#So we drop them
#Drop duplicates
df_combined = df_combined.drop_duplicates()
#check the info of the combined tables to confirm if the duplicates have been dropped
df_combined
#we notice that previously we had 15001 rows now we have 9841 rows
```

```python
#we can now merge the cell_geo dataset with our above df_combined dataset
df_combined_geo = pd.merge(df_combined,df5, how='inner', on=['Cell_ID'])
df_combined_geo.head()
```
