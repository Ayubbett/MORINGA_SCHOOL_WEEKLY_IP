**Data Analysis**

**Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.**

```python
paris = df[df['city'] == 'Paris']
paris_total = paris.groupby(['public name','date_time'])[['bluecar counter']].sum()
paris_total.head()
```

```python
#Lets sort it in descending order
paris_total1 = paris_total.sort_values('bluecar counter', ascending=False)
paris_total1.head()
```

```python
#Therefore the most popular hour for picking a car in April 2018 are two: ie 03:12 and 17:24 with both having a bluecar counter of 7.
```

**What is the most popular hour for returning cars?**

```python
#Lets first get the rows where bluecar was used
pop_hour = df[df['bluecar counter'] > 0]

pop_hour.head()
```

```python
# then look for time and counter
bluecar = pop_hour.groupby(['hour'])[['bluecar counter']].sum()
bluecar.head()
```

```python
#lets sort them in descending order to see the top most

bluecar = bluecar.sort_values('bluecar counter', ascending=False)
bluecar.head()
```

```python
# Most popular hour for returning a car is 6 with bluecar counter of 608
```
